```python
def pull_des(x):
    
    func_resp = requests.get(x)
    
    func_soup = BS(func_resp.text)
    
    func_tags = func_soup.findAll('p')
    
    return func_tags[1].text
```

```python
pull_des(fake_jobsdf.loc[0, 'apply2'])
```

```python
fake_jobsdf['description_2'] = fake_jobsdf.apply2.apply(pull_des)
fake_jobsdf
```
